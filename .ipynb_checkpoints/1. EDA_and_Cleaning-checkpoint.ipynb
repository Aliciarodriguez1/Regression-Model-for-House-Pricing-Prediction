**Project 2: Ames, Housing Model**

**Problem Statement**

To predict house pricing in Ames, Iowa for house buyers and sellers by analyzing qualitative and quantitative features creating a regression model.

**Executive Summary**

**Overview**

Kaggle which is an online community of data scientists and machine learning apprentices have shared the Ames Housing Dataset which has 81 columns of different house features.This dataset consists of 39 quantitave features and 42 qualitative features which will be analyzed separately. This model will make predictions of the house pricing by analyzing in detail qualitative and quantitaive features and their correlation between the sale price by creating a regression model to see its trend.
